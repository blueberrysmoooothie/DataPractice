2. 두 테이블을 내부 조인(buy_table.useID와 user_table.userID)한 다음, 아래의 결과와 같이 출력이 되도록 SQL 쿼리를 작성하시오.
- 1) 내부 조인한 결과에 ‘연락처’ 컬럼 추가

```python
import pandas as pd
conn = pymysql.connect(host = 'localhost', user = 'root', password = 'pokespe123!',
                      db = 'shoppingmall', charset = 'utf8')
curs = conn.cursor()
sql = """
select u.userName, b.prodName, u.addr, concat(u.mobile1, u.mobile2) as 연락처
from user_table as u
inner join buy_table as b
on u.userID = b.userID;
"""
curs.execute(sql)
rows = curs.fetchall()

call_df = pd.DataFrame(rows, columns = ['userName', 'prodName', 'addr', '연락처'])
curs.close()
conn.close()

call_df
```

- 2) userID가 KYM인 사람이 구매한 물건과 회원 정보 출력

```python
conn = pymysql.connect(host = 'localhost', user = 'root', password = 'pokespe123!',
                      db = 'shoppingmall', charset = 'utf8')
curs = conn.cursor()
sql = """
select u.userID, u.userName, b.prodName, u.addr,  concat(u.mobile1, u.mobile2) as 연락처
from user_table as u
inner join buy_table as b
on u.userID = b.userID
where u.userID = 'KYM';
"""
curs.execute(sql)
rows = curs.fetchall()

KYM_df = pd.DataFrame(rows, columns = ['userID','userName', 'prodName', 'addr', '연락처'])
curs.close()
conn.close()

KYM_df
```

- 3) 전체 회원이 구매한 목록을 회원 아이디 순으로 정렬

```python
conn = pymysql.connect(host = 'localhost', user = 'root', password = 'pokespe123!',
                      db = 'shoppingmall', charset = 'utf8')
curs = conn.cursor()
sql = """
select u.userID, u.userName, b.prodName, u.addr,  concat(u.mobile1, u.mobile2) as 연락처
from user_table as u
inner join buy_table as b
on u.userID = b.userID
order by u.userID;
"""
curs.execute(sql)
rows = curs.fetchall()

total_order_df = pd.DataFrame(rows, columns = ['userID','userName', 'prodName', 'addr', '연락처'])
curs.close()
conn.close()

total_order_df
```

- 4) 쇼핑몰에서 한 번이라도 구매한 기록이 있는 회원 정보를 회원 아이디 순으로 출력 (distinct 사용)

```python
conn = pymysql.connect(host = 'localhost', user = 'root', password = 'pokespe123!',
                      db = 'shoppingmall', charset = 'utf8')
curs = conn.cursor()
sql = """
select distinct u.userID, u.userName, u.addr
from user_table as u
inner join buy_table as b
on u.userID = b.userID
order by u.userID;
"""
curs.execute(sql)
rows = curs.fetchall()

uniqe_user_df = pd.DataFrame(rows, columns = ['userID','userName', 'addr'])
curs.close()
conn.close()

uniqe_user_df
```

- 5) 쇼핑몰 회원 중에서 주소가 경북과 경남인 회원 정보를 회원 아이디 순으로 출력

```python
conn = pymysql.connect(host = 'localhost', user = 'root', password = 'pokespe123!',
                      db = 'shoppingmall', charset = 'utf8')
curs = conn.cursor()
sql = """
select u.userID , u.userName , u.addr ,concat(u.mobile1, u.mobile2) as 연락처
from user_table u
where u.addr in ('경북', '경남')
order by u.userID;
"""
curs.execute(sql)
rows = curs.fetchall()

KyungGi_df = pd.DataFrame(rows, columns = ['userID','userName', 'addr', '연락처'])
curs.close()
conn.close()

KyungGi_df
```
