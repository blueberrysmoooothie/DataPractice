```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform
import re

# 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family = 'Malgun Gothic')
else:
    plt.rc('font', family = 'AppleGothic')
```

```python
# 지역별 pie 차트 그리기 (지역명, 남성 수, 여성 수)
def draw_pie(disctrict, male_num, female_num):
    # 문자열을 int로 변환하여 pie chart 생성
    plt.pie([int(male_num), int(female_num)],
           labels = ['남성', '여성'],
           autopct = '%.1f%%',
           startangle= 90)
    plt.title(disctrict)

# csv모듈 이용
def get_city_data_csv(city):
    # '구' 명, 남성 수, 여성 수 저장할 리스트 생성
    district_list = []
    male_num_list = []
    female_num_list = []
    
    # 지역 전체 남성 수, 여성 수 저장 리스트
    city_total = []
    
    with open('./gender.csv', encoding = 'euc-kr') as f:
        data = csv.reader(f)
        cols = next(data)
        
        # 지역 전체 자료는 가장 처음 나오므로 최초 탐색 성공시 전체 자료 저장 후 flag 변환
        total_flag = 1
        for row in data:
            if city in row[0] :
                
                # 전체 자료의 경우
                if total_flag:
                    cityname = re.split('[()]', row[0])[0].strip() 
                    city_total = [row[104].replace(',',''),row[207].replace(',','')]
                    total_flag = 0
                    continue
                    
                # 구별 자료의 경우
                district = re.split('[()]', row[0])[0].strip() 
                
                # '--구' 인지 확인 하여 처리
                if district[-1] == '구':
                    district_list.append(district)
                    male_num_list.append(row[104].replace(',',''))
                    female_num_list.append(row[207].replace(',',''))
                
    return (cityname, district_list, male_num_list, female_num_list, city_total)
   
def get_city_data_pandas(city):
    data = pd.read_csv('./gender.csv', encoding = 'euc-kr')
    
    # '구' 명, 남성 수, 여성 수 저장할 리스트 생성
    male_num_list = []
    female_num_list = []
    
    # 해당 도시에 속하는 모든 자료 슬라이싱
    city_data = data[[(city in name) for name in data.iloc[:,0]]]
    
    # 지명에 포함된 () 제거
    city_data.iloc[:,0] = [re.split('[()]', name)[0].strip() for name in city_data.iloc[:,0]]
    
    # 전체 자료 저장
    cityname = city_data.iloc[:,0].iloc[0]
    city_total = list(city_data.iloc[0,[104, 207]].str.replace(',',''))
    
    # 지명이 '--구' 인 자료 슬라이싱 후 처리
    city_data = city_data[[name[-1] == '구' for name in city_data.iloc[:,0]]].copy()
    district_list = list(city_data.iloc[:,0])
    male_num_list = list(city_data.iloc[:,104].str.replace(',',''))
    female_num_list = list(city_data.iloc[:,207].str.replace(',',''))
    
    return (cityname, district_list, male_num_list, female_num_list, city_total)
```

```python
get_city_data_pandas('대구광역시')
```

```python
get_city_data_csv('대구광역시')
```

```python
def city_gender_rate(city, data_mod = 'csv'):
    # csv
    if data_mod == 'csv':
        cityname, district_list, male_num_list, female_num_list, city_total = get_city_data_csv(city)
    # pandas
    elif data_mod == 'pandas':
        cityname, district_list, male_num_list, female_num_list, city_total = get_city_data_pandas(city)
    
    # 4열의 subplot들을 그리도록 설정
    r,c = len(district_list)//4 +1,4
    fig, axes = plt.subplots(r, c, figsize = (12,8))
    plt.suptitle(cityname + ' 구별 남녀 인구 비율')
    
    color = ['tomato', 'royalblue']
    
    # 각 구별로 pie 차트 그리기
    for i in range(len(district_list)):
        plt.subplot(r, c,i+1)
        draw_pie(district_list[i], male_num_list[i], female_num_list[i])
        
    # 전체 총합 자료 pie 차트 그리기
    plt.subplot(r, c, len(district_list)+1)
    draw_pie(cityname+' 전체', city_total[0], city_total[1])
```

```python
city_gender_rate('대구광역시')
```

```python
city_gender_rate('대구광역시', data_mod = 'pandas')
```
