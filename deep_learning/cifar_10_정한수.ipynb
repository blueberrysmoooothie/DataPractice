## CIFAR 이미지 분류
- 학습 방법 : 분류 (다중)
- 이미지 : DNN(MLP), CNN

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten
from tensorflow.keras.utils import set_random_seed, plot_model
from tensorflow.keras.datasets.cifar10 import load_data

```

```python
(train_x, train_y),(test_x, test_y) = load_data()
print(train_x.shape, train_y.shape)
print(test_x.shape, test_y.shape)
```

```python
train_x = train_x/255.0
test_x = test_x/255.0
```

```python
from sklearn.model_selection import train_test_split
s_train_x, val_x, s_train_y, val_y = train_test_split(train_x, train_y, stratify=train_y, random_state=493)
print(s_train_x.shape, s_train_y.shape)
print(val_x.shape, val_y.shape)
print(test_x.shape, test_y.shape)
```

```python
model = Sequential()

model.add(Conv2D(
    6,
    kernel_size = 3,
    padding = 'same',
    input_shape = (32,32,3)
))
model.add(MaxPool2D())

model.add(Conv2D(
    3,
    kernel_size = 3,
    padding = 'same'
))
model.add(MaxPool2D())

model.add(Flatten())
model.add(Dropout(0.1))
model.add(Dense(
    50, 
    activation='relu'
))
model.add(Dense(
    10, 
    activation='softmax'
))

model.summary()
plot_model(model,show_shapes=True)
```

```python
model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
filepath='./cifar_best_model'
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
es_cb = EarlyStopping(monitor = 'val_loss',patience = 5)
ck_cb = ModelCheckpoint(filepath = filepath,
                       monitor = 'val_loss',
                       save_best_only = True)
set_random_seed(493)

result =model.fit(
        s_train_x, s_train_y,
        epochs = 200,
        batch_size = 400,
        validation_data = (val_x, val_y),
        callbacks= [es_cb,ck_cb],
        verbose = 1
)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
for k in result.history.keys():
    plt.plot(result.epoch, result.history[k], label = k)
    
plt.legend()
plt.show()
```

```python
from keras.models import Model, load_model, save_model
filepath='./cifar_best_model'
model = load_model(filepath=filepath)
```

```python
model.evaluate(test_x, test_y)
```

```python
import numpy as np
plt.matshow(test_x[0], cmap = 'gray_r')
pred = model.predict(test_x[[0]])
print(test_y[0])
print(np.round(pred,2))
```
