```python
# ny_taxi = taxi[taxi['in_ny']] 
# ny_taxi.to_csv('./ny_taxi.csv', index=False)
ny_taxi = pd.read_csv('./ny_taxi.csv')
ny_taxi['dist'] = np.sqrt((ny_taxi.pickup_latitude - ny_taxi.dropoff_latitude)**2 +(ny_taxi.pickup_longitude - ny_taxi.dropoff_longitude)**2)
ny_taxi
```

```python
dist_ny_taxi = ny_taxi[['dist', 'fare_amount']]
dist_ny_taxi['dist'].describe()
```

```python
dist_ny_taxi['long_dist'] = np.where(dist_ny_taxi['dist']>0.5, dist_ny_taxi['dist']-0.5, 0)
dist_ny_taxi['dist'] = np.where(dist_ny_taxi['dist']<=0.5, dist_ny_taxi['dist'], 0.5)
```

```python
dist_ny_taxi
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# dist_ny_taxi['s_dist'] = dist_ny_taxi['dist'] ** 2
train_x, test_x, train_y, test_y = train_test_split(dist_ny_taxi[['dist','long_dist']], dist_ny_taxi['fare_amount'], random_state = 493)
```

```python
plt.scatter(train_x['dist'], train_y)
```

```python
train_y = np.array(train_y).reshape(-1,1)
test_y = np.array(test_y).reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(train_x, train_y)
print('w:', lr.coef_, ' b:', lr.intercept_)
```

```python
print('train score :', lr.score(train_x, train_y))
print('test score :', lr.score(test_x, test_y))
```

### 과제 2

```python
import pandas as pd
data = pd.read_csv('./diabetes.csv')
data
```

```python
import numpy as np
data.iloc[:,[1,2,3,4,5,7]] = np.where(data.iloc[:,[1,2,3,4,5,7]] == 0, np.nan, data.iloc[:,[1,2,3,4,5,7]])
data = data.dropna()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(data.drop('Outcome', axis = 1), data['Outcome'])
train_y = np.array(train_y).reshape(-1,1)
test_y = np.array(test_y).reshape(-1,1)
```

```python
logit_model = LogisticRegression()
logit_model.fit(train_x, train_y)

print('w:', logit_model.coef_, ' b:', logit_model.intercept_)
print('train score :', logit_model.score(train_x, train_y))
print('test score :', logit_model.score(test_x, test_y))
```

```python
new_data = [[0,60,72,20,60,27,0,26]]
pred_y = logit_model.predict_proba(new_data)[0][1]
print()
print(f'당뇨병 발병 확률이 {round(pred_y, 2)}%입니다.')
```
